```python
# k nearest neighbors - find via distance (euclidean), then assign majority class or take mean of points

# KNN Classification
# def knn(x: torch.tensor, points: torch.tensor, labels: torch.tensor, k: int):

#   # dist of x to every point
#   dist = torch.sqrt(torch.sum((points-x) ** 2, dim=-1))

#   # get indices of k smallest dist
#   nearest_indices = torch.topk(dist, k, largest=False).indices

#   # get labels of nearest neighbors
#   nearest_labels = labels[nearest_indices]

#   # majority vote
#   pred = torch.mode(nearest_labels).values

#   return pred

# KNN Regression (continuous prediction)
def knn(x: torch.tensor, points: torch.tensor, values: torch.tensor, k: int):

  # dist of x to every point
  dist = torch.sqrt(torch.sum((points - x) ** 2, dim=-1)) # euclidean calc

  # indices of k nearest neighbors
  nearest_indices = torch.topk(dist, k, largest=False).indices

  # get nearest values
  nearest_values = values[nearest_indices]

  # take the mean
  pred = torch.mean(nearest_values.float())

  return pred
```
