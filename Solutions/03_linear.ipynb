# 🟡 Medium: Simple Linear Layer

Implement a fully-connected linear layer: **y = xW^T + b**

### Signature
```python
class SimpleLinear:
    def __init__(self, in_features: int, out_features: int): ...
    def forward(self, x: torch.Tensor) -> torch.Tensor: ...
```

### Requirements
- `self.weight`: shape `(out_features, in_features)`, init with `randn * (1/√in_features)`
- `self.bias`: shape `(out_features,)`, init as zeros
- Both must have `requires_grad=True`
- `forward(x)` computes `x @ W^T + b`
- Do **NOT** use `torch.nn.Linear`

````python
import torch
import math
````

````python
class SimpleLinear:
  def __init__(self, in_features: int, out_features: int):
    # store dimensions
    self.in_features = in_features
    self.out_features = out_features

    # init weights and bias
    self.weight = torch.randn(out_features, in_features) * (1 / in_features**0.5) # inits randomly assigned weights, then scales weight down for training
    self.weight.requires_grad_(True) # computes gradients for learning
    self.bias = torch.zeros(out_features) # inits bias vector / neuron with a bunch of 0s
    self.bias.requires_grad_(True) # computes gradients for learning

  def forward(self, x: torch.tensor) -> torch.tensor:
    return x @ self.weight.T + self.bias # matrix multiplication, transpose
````

````python
# 🧪 Debug

layer = SimpleLinear(8, 4)
print("W shape:", layer.weight.shape)
print("b shape:", layer.bias.shape)

x = torch.rand(2, 8)
y = layer.forward(x)
print("Output shape:", y.shape)



# layer = SimpleLinear(8, 4)
# print("W shape:", layer.weight.shape)  # should be (4, 8)
# print("b shape:", layer.bias.shape)    # should be (4,)

# x = torch.randn(2, 8)
# y = layer.forward(x)
# print("Output shape:", y.shape)        # should be (2, 4)
````
